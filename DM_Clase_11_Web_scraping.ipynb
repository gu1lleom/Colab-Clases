# Web scraping

Leemos p√°ginas web para sacar datos estructurados

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
res = requests.get('https://auto.mercadolibre.com.ar/MLA-1106582850-amarok-30-tdi-v6-258cv-comfortline-4x4-aut-2022-_JM#position=1&search_layout=grid&type=item&tracking_id=36bed41c-0b26-4278-ba04-49d53f319917')

if res.status_code == 200:
    print(res.content)
```

## Beautiful Soup

```python
tree = BeautifulSoup(res.content, 'html.parser')
print(tree.prettify()[:400])
```

```python
tree.find(class_="andes-money-amount__fraction").get_text()
```

```python
tree.find(class_="andes-table")
```

```python
for row in tree.find_all(class_="andes-table__header"):
    print(row.get_text())
```

```python
for row in tree.find_all(class_="andes-table__column"):
    print(row.get_text())
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome('chromedriver', options=chrome_options)
```

```python
driver.get("https://mudafy.com.ar/departamento/av-scalabrini-ortiz-1472-8-33-departamento-en-venta-1evg")
```

```python
driver.find_element(By.CLASS_NAME, "listing-title__title").text
```

```python
driver.find_element(By.CLASS_NAME, "prices__price").text
```

```python
for e in driver.find_elements(By.CLASS_NAME, "detail"):
    print(e.text)
```

```python
driver.find_element(By.CLASS_NAME, "listing-description__text").text
```
